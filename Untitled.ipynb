```python
from flask import Flask, request, render_template_string, redirect
from flask_cloudflared import run_with_cloudflared
import cv2
import numpy as np
import face_recognition
import base64
import pickle
import os
import time
import csv
from datetime import datetime
import pytz
from google.colab import drive

# ================= MOUNT DRIVE =================
drive.mount('/content/drive')

SAVE_PATH = "/content/drive/MyDrive/enrolled_faces.pkl"
LOG_PATH = "/content/drive/MyDrive/access_log.csv"

app = Flask(__name__)
run_with_cloudflared(app)

# ================= GLOBAL VARIABLES =================
enrolled_faces = []
last_image_b64 = ""
detect_mode = False
last_result = "READY"
result_timestamp = 0

# ================= LOAD SAVED FACES =================
if os.path.exists(SAVE_PATH):
    with open(SAVE_PATH, 'rb') as f:
        enrolled_faces = pickle.load(f)

# ================= CREATE LOG FILE IF NOT EXISTS =================
if not os.path.exists(LOG_PATH):
    with open(LOG_PATH, 'w', newline='') as f:
        writer = csv.writer(f)
        writer.writerow(["Timestamp", "Status"])

# ================= MALAYSIA TIME LOG FUNCTION =================
def log_access(status):
    malaysia = pytz.timezone("Asia/Kuala_Lumpur")
    now = datetime.now(malaysia).strftime("%d-%m-%Y %H:%M:%S")

    with open(LOG_PATH, 'a', newline='') as f:
        writer = csv.writer(f)
        writer.writerow([now, status])

# ================= HTML =================
HTML_PAGE = '''
<!DOCTYPE html>
<html>
<head>
<title>Smart Facial Door</title>

<style>
body { background:#0f172a; color:white; text-align:center; font-family:Arial; }
.container { width:90%; margin:auto; margin-top:30px; padding:20px; background:#1e293b; border-radius:15px; }
.status { padding:15px; border-radius:10px; font-size:20px; margin:15px 0; }
.authorized { background:#16a34a; }
.unauthorized { background:#dc2626; }
.ready { background:#facc15; color:black; }
button { padding:12px 25px; font-size:16px; border:none; border-radius:8px; margin:10px; cursor:pointer; }
.btn-detect { background:#3b82f6; color:white; }
.btn-enroll { background:#9333ea; color:white; }
img { width:50%; border-radius:12px; border:4px solid #3b82f6; margin-top:15px; }

.log-box {
    margin-top:20px;
    background:#0f172a;
    padding:15px;
    border-radius:10px;
    height:200px;
    overflow-y:scroll;
    text-align:left;
    font-size:14px;
}
</style>

<script>
function refreshImage() {
    fetch('/image')
    .then(response => response.text())
    .then(data => {
        if (data.length > 0) {
            document.getElementById("camera").src =
                "data:image/jpeg;base64," + data;
        }
    });
}

function refreshStatus() {
    fetch('/view_status')
    .then(response => response.text())
    .then(data => {
        let box = document.getElementById("statusBox");

        if (data === "AUTHORIZED") {
            box.className = "status authorized";
            box.innerHTML = "ACCESS GRANTED";
        }
        else if (data === "UNAUTHORIZED") {
            box.className = "status unauthorized";
            box.innerHTML = "ACCESS DENIED";
        }
        else {
            box.className = "status ready";
            box.innerHTML = "SYSTEM READY";
        }
    });
}

function refreshLogs() {
    fetch('/logs')
    .then(response => response.text())
    .then(data => {
        document.getElementById("logBox").innerHTML = data;
    });
}

setInterval(refreshImage, 1000);
setInterval(refreshStatus, 300);
setInterval(refreshLogs, 1000);
</script>

</head>
<body>

<div class="container">
<h1>SMART FACIAL DOOR SYSTEM</h1>

<img id="camera" src="data:image/jpeg;base64,{{ image }}">

<div id="statusBox" class="status ready">SYSTEM READY</div>

<p>Enrolled Faces: {{ count }}</p>

<form action="/start_detect" method="post">
<button class="btn-detect">START DETECTION</button>
</form>

<form action="/enroll" method="post">
<button class="btn-enroll">ENROLL FACE</button>
</form>

<h2>Access Logs (Malaysia Time)</h2>
<div id="logBox" class="log-box">Loading logs...</div>

</div>
</body>
</html>
'''

# ================= VIEW =================
@app.route('/view')
def view():
    return render_template_string(
        HTML_PAGE,
        count=len(enrolled_faces),
        image=last_image_b64
    )

# ================= IMAGE =================
@app.route('/image')
def image():
    return last_image_b64

# ================= STATUS =================
@app.route('/view_status')
def view_status():
    global last_result, result_timestamp

    if last_result in ["AUTHORIZED", "UNAUTHORIZED"]:
        if time.time() - result_timestamp > 3:
            last_result = "READY"

    return last_result

# ================= LOG DISPLAY =================
@app.route('/logs')
def logs():
    if not os.path.exists(LOG_PATH):
        return "No logs yet."

    with open(LOG_PATH, 'r') as f:
        lines = f.readlines()

    lines = lines[::-1]
    return "<br>".join(lines[:20])

# ================= START DETECT =================
@app.route('/start_detect', methods=['POST'])
def start_detect():
    global detect_mode
    detect_mode = True
    return redirect("/view")

# ================= ENROLL =================
@app.route('/enroll', methods=['POST'])
def enroll():
    global last_image_b64

    if not last_image_b64:
        return redirect("/view")

    img_bytes = base64.b64decode(last_image_b64)
    img = cv2.imdecode(np.frombuffer(img_bytes, np.uint8), cv2.IMREAD_COLOR)
    rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    encodings = face_recognition.face_encodings(rgb)

    if len(encodings) > 0:
        enrolled_faces.append(encodings[0])
        with open(SAVE_PATH, 'wb') as f:
            pickle.dump(enrolled_faces, f)

    return redirect("/view")

# ================= RECOGNITION =================
@app.route('/recognize', methods=['POST'])
def recognize():
    global last_image_b64, detect_mode, last_result, result_timestamp

    file = request.files['image'].read()
    last_image_b64 = base64.b64encode(file).decode('utf-8')

    if not detect_mode:
        return "IDLE", 200

    img = cv2.imdecode(np.frombuffer(file, np.uint8), cv2.IMREAD_COLOR)
    rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    small = cv2.resize(rgb, (0, 0), fx=0.5, fy=0.5)

    locations = face_recognition.face_locations(small)
    encodings = face_recognition.face_encodings(small, locations)

    detect_mode = False

    if len(encodings) > 0 and enrolled_faces:
        distances = face_recognition.face_distance(enrolled_faces, encodings[0])
        best_match = np.argmin(distances)

        # ðŸ”¥ UPDATED THRESHOLD
        if distances[best_match] < 0.45:
            last_result = "AUTHORIZED"
            result_timestamp = time.time()
            log_access("AUTHORIZED")
            return "MATCH FOUND", 200

    last_result = "UNAUTHORIZED"
    result_timestamp = time.time()
    log_access("UNAUTHORIZED")
    return "UNKNOWN", 401

# ================= RUN =================
app.run()

```
